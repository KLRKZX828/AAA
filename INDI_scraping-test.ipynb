```python
import os
import json
import pandas as pd

# Pfad zum Hauptordner, in dem sich die Account-Ordner befinden
main_directory = "/Users/pressmann/INDI_TT_Scraping"

# Initialisierung der Listen für die gesammelten Daten
like_counts, share_counts, comment_counts, durations = [], [], [], []

# Durch den Hauptordner iterieren und in jedem Account-Ordner die JSON-Dateien verarbeiten
for account_folder in os.listdir(main_directory):
    account_path = os.path.join(main_directory, account_folder)
    
    # Sicherstellen, dass es sich um ein Verzeichnis handelt
    if os.path.isdir(account_path):
        for filename in os.listdir(account_path):
            if filename.endswith(".json"):
                with open(os.path.join(account_path, filename), 'r') as file:
                    data = json.load(file)
                    
                    # Daten sammeln
                    like_counts.append(data.get('like_count', 0))
                    share_counts.append(data.get('repost_count', 0))
                    comment_counts.append(data.get('comment_count', 0))
                    durations.append(data.get('duration', 0))

# Berechnungen für die Statistiken
result = {
    "Anzahl Videos": len(durations),
    "Durchschnitt Likes": sum(like_counts) / len(like_counts) if like_counts else 0,
    "Minimum Likes": min(like_counts) if like_counts else 0,
    "Maximum Likes": max(like_counts) if like_counts else 0,
    "Durchschnitt Shares": sum(share_counts) / len(share_counts) if share_counts else 0,
    "Minimum Shares": min(share_counts) if share_counts else 0,
    "Maximum Shares": max(share_counts) if share_counts else 0,
    "Durchschnitt Kommentare": sum(comment_counts) / len(comment_counts) if comment_counts else 0,
    "Minimum Kommentare": min(comment_counts) if comment_counts else 0,
    "Maximum Kommentare": max(comment_counts) if comment_counts else 0,
    "Durchschnitt Videodauer": sum(durations) / len(durations) if durations else 0,
    "Minimum Videodauer": min(durations) if durations else 0,
    "Maximum Videodauer": max(durations) if durations else 0,
    "Summe Videodauer": sum(durations) if durations else 0,
}

# Resultate anzeigen
print(result)

# Optional: Speichern in eine CSV-Datei
df = pd.DataFrame([result])
df.to_csv('statistiken_videos.csv', index=False)
```

```python
import os
import json
import pandas as pd

main_directory = '/Users/pressmann/INDI_TT_Scraping'

# Liste für Accounts mit 2 oder weniger Videos und TikTok Handle
accounts_with_few_videos = []

# Durch jeden Account-Ordner im Hauptordner iterieren und die Anzahl der JSON-Dateien zählen
for account_folder in os.listdir(main_directory):
    account_path = os.path.join(main_directory, account_folder)
    
    if os.path.isdir(account_path):
        # Zähle JSON-Dateien im Account-Ordner
        video_count = sum(1 for filename in os.listdir(account_path) if filename.endswith(".json"))
        
        # Wenn 2 oder weniger Videos, Account zur Liste hinzufügen
        if video_count <= 2:
            # Extrahieren des TikTok Handle (aus einer JSON-Datei im Ordner, wenn vorhanden)
            tiktok_handle = None
            for filename in os.listdir(account_path):
                if filename.endswith(".json"):
                    with open(os.path.join(account_path, filename), 'r') as file:
                        data = json.load(file)
                        tiktok_handle = data.get('uploader_url', 'Kein Handle gefunden')
                    break  # Nur die erste JSON-Datei prüfen
            
            # Account und Handle zur Liste hinzufügen
            accounts_with_few_videos.append((account_folder, tiktok_handle))

# Ergebnisse in DataFrame speichern und in CSV exportieren
df = pd.DataFrame(accounts_with_few_videos, columns=['Account Name', 'TikTok Handle'])
csv_path = 'suspicious_accounts.csv'
df.to_csv(csv_path, index=False)

print(f"Die CSV-Datei wurde erstellt: {csv_path}")
```
